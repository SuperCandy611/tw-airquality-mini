```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Microsoft JhengHei"
plt.rcParams["axes.unicode_minus"] = False

from src.load_data import load_csv
from src.clean_data import clean
```

```python
import numpy as np

np.random.seed(42)

df = pd.DataFrame({
    "date":    pd.date_range("2025-01-01", periods=10, freq="D"),
    "station": "Taipei",
    "pm25":    np.random.randint(10, 81, size=10),
})

df
```

```python
fig, ax = plt.subplots(figsize=(9, 4))

ax.plot(df["date"], df["pm25"], marker="o", linewidth=1.8, color="#4c8fbd")
ax.set_title("PM2.5 Demo — Taipei (10 days)", fontsize=13, pad=10)
ax.set_xlabel("Date", fontsize=11)
ax.set_ylabel("PM2.5 (µg/m³)", fontsize=11)
ax.set_ylim(0, 90)
ax.grid(axis="y", linestyle=":", linewidth=0.6, alpha=0.7)
fig.autofmt_xdate()
fig.tight_layout()

plt.show()
```

```python
from pathlib import Path

out = Path("..") / "reports" / "pm25_demo.png"
fig.savefig(out, dpi=150)
print(f"Saved → {out.resolve()}")
```
